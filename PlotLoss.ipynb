```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
with open('full_test1/loss_history.pkl','rb') as f:
    data = pickle.load(f)

overall_loss = data[0]
value_loss = data[1]
policy_loss = data[2]

x = np.linspace(1,30,30)
ax = plt.axes()
plt.plot(x,overall_loss,label='overall')
plt.plot(x,value_loss,label='value')
plt.plot(x,policy_loss,label='policy')
plt.legend()
```
